# Validação do ETL: Busca de Dados no MongoDB
Este notebook tem como objetivo conectar-se ao banco de dados e validar que as coleções criadas no processo de ETL possuem dados inseridos corretamente.

```python
import pymongo
import json
import os
from bson import ObjectId
from dotenv import load_dotenv

# Carrega as variáveis de ambiente do backend para proteger a senha
load_dotenv("../backend/.env")
mongo_url = os.getenv("MONGO_URL")

# Conexão com MongoDB
client = pymongo.MongoClient(mongo_url)

# Banco de dados
db = client["siest_db"]
```

```python
colecoes = [
    "casos_geolocalizados",
    "clima",
    "risco_inundacao",
    "vulnerabilidade_habitacional"
]

print("="*60)
print("   VALIDAÇÃO DO ETL: BUSCA DE DADOS INSERIDOS NO MONGODB")
print("="*60)

for col_name in colecoes:
    col = db[col_name]
    count = col.count_documents({})
    print(f"\n🚀 Coleção: {col_name}")
    print(f"Total de documentos (registros): {count}")
    
    if count > 0:
        print("Exemplo de um documento (amostra):")
        exemplo = col.find_one()
        # Tratamento simples para printar ObjectId do MongoDB
        if isinstance(exemplo.get('_id'), ObjectId):
            exemplo['_id'] = str(exemplo['_id'])
        
        # Print formatado do dicionário
        print(json.dumps(exemplo, indent=4, ensure_ascii=False, default=str))
    else:
        print("⚠️  Nenhum dado encontrado nesta coleção.")
        
print("\n" + "="*60)
print("   VALIDAÇÃO CONCLUÍDA")
print("="*60)
```
